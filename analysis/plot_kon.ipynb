```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
txp_init = [2/150.0,20/150.0] #per sec
sc = [0,-0.06]
```

```python
def k_on(s,k0=0.0,s0=0.05,mu=25.0):
    # k0=0.01
    return k0+s0*(1. - np.tanh(mu * (s + 0.05)))

sc_values = np.linspace(-0.15,0.1,100)
plt.figure(figsize=(4,3), dpi=100)
plt.plot(sc, np.array(txp_init)*60, 'o', label='expt')
# plt.plot(sc_values, k_on(sc_values,k0=0.05)*60, label="sim fit")
# plt.plot(sc_values, k_on(sc_values,k0=0.04)*60, label="pred high txp")
plt.plot(sc_values, k_on(sc_values,k0=0.0,s0=0.1)*60, label="pred low txp")
plt.xlabel("Supercoiling density")
plt.ylabel("Transcription initiaion rate $k_{on}^{pol}$ (1/min)")

plt.ylim(0,0.2*60)
plt.xlim(-0.15,0.04)

plt.legend()
```
